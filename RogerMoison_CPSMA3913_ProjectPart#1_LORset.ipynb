# Powerset

The powerset is definately more painful then the rest of this still try to figure it out honestly.

```
def DPpowerset(s):
  powerset = [[]]
  for i in s:
    for sett in powerset:
      powerset = powerset + [sett +[i]]
  return powerset
  
```
